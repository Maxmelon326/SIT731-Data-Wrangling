### Introduction

BMI (Body Mass Index) is a commonly used measure of weight relative to height, calculated as weight in kilograms divided by the square of height in meters. This report is an exploration of the initial use of python's jupyter notebook to measure and graphically display BMI values for different criteria for different samples, and to explore the limitations of the use of BMI values. 


### Case Study of BMI
BMI is widely used in modern life and is commonly used by people to simply assess whether they are healthy or not. According to Wikipedia's definition, there are multiple ways to calculate BMI and there are different standard ranges in different countries. Traditinoal BMI version is BMI = [weight / ((height / 100) ** 2). In recent years, several studies have proposed new BMI algorithms that aim to more accurately reflect individual body fat content, and the New BMI = New_BMI= [1.3 * (weight / ((height / 100) ** 2.5)). In order to explore the pros and cons of BMI, firstly, some Chinese samples were selected for this experiment, the sample attributes contain name, height (cm) and weight (kg), measure BMI and new BMI, the samples are as follows:

```python
#The Samples data are Chinese, then we can measure the healthy status based on WHO BMI ranges. 
names = ["Jane", "Star", "Celine", "Belinda", "Rokee", "Sam","Aj"]
heights = [160,165,163,135, 180, 175, 171 ] # in centimetres
weights = [74, 58, 60, 50, 75,108,65 ] # in kilograms
```

```python
#According to the Wikipedia definitions of BMI and New BMI(exponent of 2.5), we can calculate BMI & New_BMI
BMI = [weight / ((height / 100) ** 2) for height, weight in zip(heights, weights)]
New_BMI= [1.3 * (weight / ((height / 100) ** 2.5)) for height, weight in zip(heights, weights)]
```

```python
#State sample's BMI and New_BMI
#The WHO regards an adult BMI of less than 18.5 as underweight and possibly indicative of malnutrition,an eating disorder, or other health problems, 
#while a BMI of 25 or more is considered overweight and 30 or more is considered obese.
#In addition to the principle, international WHO BMI cut-off points (16, 17, 18.5, 25, 30, 35 and 40)

# Calculate BMI and New_BMI labels
for i, (name, bmi, new_bmi) in enumerate(zip(names, BMI, New_BMI)):
    # Calculate BMI label
    if bmi < 18.5:
        label = "Underweight"
    elif 18.5 <= bmi <= 24.9:
        label = "Normal Weight"
    elif 25.0 <= bmi <= 29.9:
        label = "Pre-obesity"
    elif 30.0 <= bmi <= 34.9:
        label = "Obesity class I"
    elif 35.0 <= bmi <= 39.9:
        label = "Obesity class II"
    else:
        label = "Obesity class III"
        
    # Calculate New_BMI label
    if new_bmi < 18.5:
        new_label = "Underweight"
    elif 18.5 <= new_bmi <= 24.9:
        new_label = "Normal Weight"
    elif 25.0 <= new_bmi <= 29.9:
        new_label = "Pre-obesity"
    elif 30.0 <= new_bmi <= 34.9:
        new_label = "Obesity class I"
    elif 35.0 <= new_bmi <= 39.9:
        new_label = "Obesity class II"
    else:
        new_label = "Obesity class III"
    
    print(f"{name} has BMI of {bmi:.2f} which is {label}. The new BMI index is {new_bmi:.2f} which is {new_label}.")
```

```python
#Draw the bar graph to display samples' BMI and New BMI

import matplotlib.pyplot as plt

# Set the figure size
fig, ax = plt.subplots(figsize=(10, 6))

# Define colors for each category
category_colors = {
    'Underweight': 'blue', 
    'Normal Weight': 'green', 
    'Pre-obesity': 'orange',
    'Obesity class I': 'pink', 
    'Obesity class II': 'purple', 
    'Obesity class III': 'red'
}

# Plot BMI and New_BMI bars with category-specific colors
bars1 = ax.bar([i - 0.2 for i in range(len(names))], BMI, width=0.4, color=[category_colors[label] for label in ['Underweight', 'Normal Weight', 'Normal Weight', 'Pre-obesity', 'Normal Weight', 'Obesity class II', 'Normal Weight']], label='BMI')
bars2 = ax.bar([i + 0.2 for i in range(len(names))], New_BMI, width=0.4, color=[category_colors[label] for label in ['Normal Weight', 'Normal Weight', 'Normal Weight', 'Obesity class I', 'Normal Weight', 'Obesity class I', 'Normal Weight']], label='New BMI')

# Set x-axis ticks and labels
ax.set_xticks(range(len(names)))
ax.set_xticklabels(names)

# Set plot attributes
ax.set_title("BMI and New BMI Comparison")
ax.set_xlabel("Names")
ax.set_ylabel("BMI / New BMI")

# Display legend within the plot
ax.legend(handles=[plt.Rectangle((0, 0), 1, 1, color=color, ec="k") for color in category_colors.values()],
          labels=category_colors.keys(), loc='upper left', bbox_to_anchor=(1.0, 1.0))

# Show the plot
plt.show()

```

### Pros and Cons of BMI Index 
BMI is an internationally recognized indicator that can measure a person's health condition. Because of the simple calculation method, people can easily calculate and have a rough idea of their health condition. Therefore, it is widely used. However, through the above samples, we can intuitively see that there are obvious drawbacks in the calculation of BMI.

First of all, BMI only takes into account the two dimensions of weight and height, without considering the age and nationality of the sample. For example, there are different BMI standards for adults and non-adults, and different BMI standards for Asian and European countries, so if people want to have a fairer understanding of whether their BMI is standardized or not, they need to get a BMI standard that is more in line with their own age and nationality to make comparisons, or else they are prone to bias.

Secondly, in terms of individual stratification, as shown in the samples, Jane is underweight under the BMI standard, while she is Normal weight under the new BMI standard, Belinda is
Pre-obesity and under the new BMI standard is Obesity class I. Sam is Obesity class II under the BMI standard and under the new BMI standard is Obesity class I. The different results may be misleading for individuals to take measures, and each individual needs more medical checkups, such as muscle mass, bone density, and other more dimensions to determine the health of the body.

Finally, WHO measures nutritional status around the world through the BMI index, which may have an impact on aid programs, policy making, etc. around the world. Therefore, the scientific validity of the BMI calculation method and standard needs to be verified in the long run.

### Conclusion
This exploration of BMI and the new BMI algorithm reveals potential benefits and limitations. Traditional BMI has long served as a standard measure for body mass, but the new BMI algorithm, considering a power of 2.5 in height, aims for improved accuracy.

The case study on Chinese samples highlights discrepancies in weight category classifications. The study underscores the importance of considering cultural and regional differences in BMI standards. Despite limitations, both BMI and the new algorithm offer quick assessments of weight status, serving as valuable tools for initial health evaluations.

As research evolves, future BMI algorithms may enhance accuracy by incorporating additional factors. While widely used, BMI's limitations emphasize the need for a nuanced and individualized approach to health assessment.
